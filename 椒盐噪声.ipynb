```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from skimage import util
from skimage import io
import skimage
import random
#添加椒盐噪声
def myPepper(img):
    tmp = np.copy(img)
    weight,height = tmp.shape[0],tmp.shape[1]
    for i in range(weight):
        for j in range(height):
            if (random.randint(0,100)<5): tmp[i][j] = random.choice((0,255))
    return tmp


#读写图片
img = cv.imread("C:\\Users\\XRUI\\Desktop\\code9.png")
img_grey = cv.imread("C:\\Users\\XRUI\\Desktop\\code9.png",cv.IMREAD_GRAYSCALE)
img_noise = myPepper(img)
img_gnoise = myPepper(img_grey)
cv.imwrite("C:\\Users\\XRUI\\Desktop\\code9_pepper.png", img_noise)

cv.imwrite("C:\\Users\\XRUI\\Desktop\\code9_gpepper.png", img_gnoise)
#实验1，计算图像大小，灰度平均值，方差，标准差，相关系数，协方差矩阵
shape = img_grey.shape
mean = np.mean(img_grey)
var = np.var(img_grey)
std = np.std(img_grey)
corr = np.corrcoef(img_grey, img_gnoise)
cov = np.cov(img_grey, img_gnoise)

img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_noise = cv.cvtColor(img_noise, cv.COLOR_BGR2RGB)


# #显示两幅图像
# cv.namedWindow("Origin")
# cv.namedWindow("Pepper")
# cv.imshow("Origin", img)
# cv.imshow("Pepper", img_noise)
# cv.waitKey(0)
# cv.destroyAllWindows() 
plt.figure(figsize=(16,16), dpi=75)
plt.subplot(221)
plt.imshow(img)
plt.subplot(222)
plt.imshow(img_noise)
plt.subplot(223)
plt.imshow(img_grey, 'gray')
plt.subplot(224)
plt.imshow(img_gnoise, 'gray')
plt.show()
```
