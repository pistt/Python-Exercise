```python
import random
import math
import numpy as np
Neural = np.ndarray(36).reshape(6,6)
np.set_printoptions(precision = 1, suppress = False)
theta = []
E = []
#初始化偏置和误差
for i in range(6):
    E.append(0)
    if i > 2:
        theta.append(random.uniform(-0.5,0.5))
    else:
        theta.append(0)

#初始化各个神经元间的连接
#邻接矩阵有向图
for i in range(6):
    for j in range(6):
        if ((i == 0 or i == 1 or i == 2) and (j == 3 or j == 4)) or ((i == 3 or i == 4) and (j == 5)):
            Neural[i][j] = random.uniform(-1.0,1.0)
        elif i == j:
            Neural[i][j] = 0
        else:
            Neural[i][j] = 0x3f3f3f3f
            
#初始化输入。对角线的值为本行输出
Neural[0][0] = 1
Neural[1][1] = 0
Neural[2][2] = 1

#设置学习率，和循环次数
times = 1000
learn = 50/times

for n in range(times):
    #计算输出
    for i in range(6):
        for j in range(6):
            if i !=j and Neural[i][j] < 0x3f3f3f3f:
                Neural[j][j] += Neural[i][i] * Neural[i][j]#（i，j）有边，j的无偏置总输入：输入*权重
        Neural[i][i] = Neural[i][i] + theta[i]#偏置后总输入
        #该层输出
        if i > 2:
            Neural[i][i] = 1 / (1 + math.exp(-Neural[i][i]))
    
    #反向误差传播
    for i in range(5,2, -1):
        sum = 0
        for j in range(5,-1,-1):
            if i != j and Neural[i][j] < 0x3f3f3f3f:
                sum += Neural[i][j] * E[j]
        if i == 5:
            sum = 1 - Neural[i][i]
        E[i] = Neural[i][i] * (1 - Neural[i][i]) * sum
        
    #权重与偏置调整
    for i in range(3,6):
        for j in range(6):
            if i != j and Neural[i][j] < 0x3f3f3f3f:
                Neural[i][j] = Neural[i][j] + learn * Neural[i][i] * E[j]
        theta[i] = theta[i] + learn * E[i]
    print("The",n + 1,"times Neural6 output:", Neural[5][5])
    if (1.0 - Neural[5][5] < 0.1):
        break
        
print("Neural output:")
for i in range(6):
    print("Neural",i + 1, "output:",Neural[i][i])

print("Neural Errors:")
for i in range(6):
    print("Neural", i + 1, ":", E[i])
    
print("Neural biased:")
for i in range(6):
    print("Neural", i + 1 ,":", theta[i])
    
print("Network weight:")
for i in range(6):
    for j in range(6):
        if i != j and Neural[i][j] < 0x3f3f3f3f:
            print("W",i + 1, j + 1, ": ", Neural[i][j], sep = "")
```
