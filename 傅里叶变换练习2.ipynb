```python
import cv2 as cv
import numpy as np
# from skimage import util
# from skimage import io
# import skimage
import matplotlib.pyplot as plt
def plot(img, fft):
    plt.subplot(121)
    plt.imshow(img,'gray')
    plt.title('original')
    plt.subplot(122)
    plt.imshow(fft,'gray')
    plt.title('center')
    return None
    
    
#实验2，生成图像并进行FFT处理
#（1）
img = cv.imread("C:\\Users\\XRUI\\Desktop\\NewFolder\\Chrysanthemum.jpg", cv.IMREAD_GRAYSCALE)
img = np.zeros([256,256], dtype = 'uint8')
img[64:192,112:144] = 255
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
s1 = np.log(abs(fshift) + 1)
plot(img,s1)
```

```python
#(2)
img2 = np.copy(img)
for i in range(1,256,2):
    for j in range(1,256,2):
        img2[i,j] = -img[i,j]
f2 = np.fft.fft2(img2)
f2shift = np.fft.fftshift(f2)
s2 = np.log(abs(f2shift) + 1)
plot(img2,s2)
```

```python
#(3)
img3 = np.rot90(img)
f3 = np.fft.fft2(img3)
f3shift = np.fft.fftshift(f3)
s3 = np.log(abs(f3shift) + 1)
plot(img3,s2)
```

```python
#(4)
img4 = img + img3
f4 = np.fft.fft2(img4)
f4shift = np.fft.fftshift(f4)
s4 = np.log(abs(f4shift) + 1)
plot(s1, s4)
```

```python
#(5)
img5 = img2 + img3
f5 = np.fft.fft2(img5)
f5shift = np.fft.fftshift(f5)
s5 = np.log(abs(f5shift)+1)
plot(img5, s5)
```
