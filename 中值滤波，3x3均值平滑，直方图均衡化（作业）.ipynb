```python
import numpy as np
import matplotlib.pyplot as pl
image = np.array(
        [[1,7,1,8,1,7,1,1],
         [1,1,1,5,1,1,1,1],
         [1,1,5,5,5,1,1,7],
         [1,1,5,5,5,1,8,1],
         [8,1,1,5,1,1,1,1],
         [8,1,1,5,1,1,8,1],
         [1,1,1,5,1,1,1,1],
         [1,7,1,8,1,7,1,1]])
#median filtering
median = np.array(image)
for i in range(1,7):
    for j in range(1,7):
        median[i][j] = sorted(image[i-1:i+2,j-1:j+2].flatten())[4]
print("median filtring:\n",median)

#3x3 neighbourhood average
average = np.array(image)
for i in range(1,7):
    for j in range(1,7):
        average[i][j] = np.mean(image[i-1:i+2,j-1:j+2].flatten())
print("\n3x3 neighbourhood average\n",average)

#Histogram equlization
rk = np.array([0.195, 0.160, 0.147, 0.106, 0.073, 0.056, 0.049, 0.041,
          0.037, 0.031, 0.027, 0.023, 0.019, 0.017, 0.012, 0.007])
length = rk.size - 1
sk = [0.195]
for i in rk[1:]:
    sk.append(sk[-1] + i)
sk = [int(round(length*i)) for i in sk]
Psk = [0 for i in range(length + 1)]
for i in range(len(rk)):
    Psk[int(sk[i])] += rk[i]
    
#print the new grey level
print("\nEach grey level's Psk is:\n(grey level, Psk)")
for i in range(len(rk)):
    print("({0:.4f}, {1:.4f})".format(i/16, Psk[i]))
    
#origin image
for i in range(len(rk)):
    pl.vlines(i/16, 0, rk[i], 'r', '--', label='example')
pl.title("histogram of origin image")
pl.xlabel("rk")
pl.ylabel("P(rk)")
pl.xlim(-0.1,1.1)
pl.ylim(0.0, 0.2)
pl.show()
#histogram image
pl.subplot(1,1,1)
for i in range(len(rk)):
    pl.vlines(i/16, 0, Psk[i], 'r', '--', label='example')
pl.title("histogram of new image")
pl.xlabel("rk_new")
pl.ylabel("P(rk_new)")
pl.xlim(-0.1,1.1)
pl.ylim(0,0.2)
pl.show()
```
