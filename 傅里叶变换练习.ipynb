```python
import cv2 as cv
import numpy as np
# from skimage import util
# from skimage import io
# import skimage
import matplotlib.pyplot as plt
img = cv.imread("C://Users//XRUI//Desktop//NewFolder/t1.jpg", cv.IMREAD_GRAYSCALE)
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
s = np.log(abs(fshift) + 1)
plt.subplot(133)
plt.imshow(s, 'gray')
plt.title("fshift")

plt.subplot(131)
plt.imshow(img)
plt.title("img")

plt.subplot(132)
plt.imshow(np.log(abs(f) + 1), 'gray')
plt.title("f")

```
